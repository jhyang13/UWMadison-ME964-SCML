```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

### Problem 2, (1): Solve for x1(t) and x2(t)

```python
# Parameters
m = 1
k = 10
c = 1
```

```python
# Define the system of equations
def equations(t, y):
    x1, v1, x2, v2 = y
    dx1dt = v1
    dx2dt = v2
    dv1dt = (c * (v2 - v1) + k * (x2 - x1)) / m
    dv2dt = (c * (v1 - v2) + k * (x1 - x2)) / m
    return [dx1dt, dv1dt, dx2dt, dv2dt]

# Initial conditions: x1(0) = 0, v1(0) = 0, x2(0) = 1, v2(0) = 0
initial_conditions = [0, 0, 1, 0]

# Time span for the solution
t_span = (0, 10)
t_eval = np.linspace(0, 10, 500)

# Solve the system
solution = solve_ivp(equations, t_span, initial_conditions, t_eval=t_eval)
```

```python
# Plot the results
plt.plot(solution.t, solution.y[0], label='x1(t)')
plt.plot(solution.t, solution.y[2], label='x2(t)')
plt.xlabel('Time t')
plt.ylabel('Displacement')
plt.title('Displacements x1(t) and x2(t)')
plt.legend()
plt.grid(True)
plt.show()
```
